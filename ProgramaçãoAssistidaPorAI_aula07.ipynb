```python
# Número Primo Refatorado
import math

def eh_primo(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
# Testando a função com alguns exemplos
testes = [2, 3, 4, 7, 10, 13, 15, 17]
for num in testes:
    resultado = eh_primo(num)
    print(f'O número {num} é primo? {resultado}')
```
